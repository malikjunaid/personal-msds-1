```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sea_ice.csv")
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Arctic'].plot(figsize=(20,5),title="Arctic")
plt.show()
```

```python
df['Antarctica'].plot(figsize=(20,5),title="Antarctica")
plt.show()
```

```python
df['Arctic'].plot(figsize=(20,5))
df['Antarctica'].plot(figsize=(20,5))
plt.title("Arctic VS Antarctica")
plt.show()
```

## The QQ (Quantile Quantile) plot

```python
import scipy.stats
import pylab
```

```python
plt.figure(figsize=(10,7))
scipy.stats.probplot(df['Antarctica'],plot=pylab)
pylab.show()
```

```python
plt.figure(figsize=(10,7))
scipy.stats.probplot(df['Arctic'],plot=pylab)
pylab.show()
```
