# Simple Linear Regression. Minimal Example

## Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Generate Random input data to train the model.

```python
observations = 1000
xs = np.random.uniform(low=-10,high=10,size=(observations,1))
zs = np.random.uniform(low=-10,high=10,size=(observations,1))

X = np.column_stack((xs,zs))
print(X.shape)
```

## Create the targets we will aim at

```python
noise = np.random.uniform(-1,1,(observations,1))

targets = 2*xs + 3*zs +5 + noise

print (targets.shape)
```

## Plot the training data

```python
shape = targets.shape
targets = targets.reshape(observations,) # Just for plotting purpose

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111,projection='3d')# Method allowing us for 3d plotting.
ax.scatter(xs,zs,targets) # Inserting data to the plot method

# Setting up the labels of the axis
ax.set_xlabel('xs')
ax.set_ylabel('ys')
ax.set_zlabel('targets')

ax.view_init(azim=-90) # azim defines the angle at which to show the 3d plot.
plt.show()

# reshaping the targets back to the original shape
targets = targets.reshape(shape)
```

## Model parameter initialization

```python
init_range = 0.1  # Initializing the weights with uniform distribution within the range.

weights = np.random.uniform(-init_range,init_range,size = (2,1))# A column vector for weights.
biases = np.random.uniform(-init_range,init_range,size = 1) 
learning_rate = 0.02
print (weights)
print (biases)
print (learning_rate)
```

## Train the machine learning model

```python
for i in range(100):
    outputs = np.dot(X,weights) + biases 
    deltas = outputs - targets 
    loss = np.sum(deltas ** 2) / 2 / observations
    print (loss)
    deltas_scaled = deltas/observations
    weights = weights - learning_rate * np.dot(X.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
    
```

```python
print (weights)
print (biases)
```

```python
plt.plot(outputs,targets)
plt.ylabel('Targets')
plt.xlabel('Predictions')
plt.show()
```
