```python
import pandas as pd
```

```python
pd.set_option("display.max_rows",50)
pd.set_option("display.max_columns",81)
```

```python
df = pd.read_csv("../datasets/housing-prices-dataset/train.csv", index_col='Id')
```

```python
df.head()
```

```python
df.info()
```

### For ease of understanding dropping all Non-Numeric columns

```python
object_columns = [column for column in df.columns if df[column].dtype == 'object']
```

```python
print ("Columns to drop\n")
for column in object_columns:
    print ("{}".format(column.ljust(50)),end=' ')
```

### Dropping columns

```python
df = df[list(set(df.columns)-set(object_columns))]
```

```python
df.head()
```

```python
df.info()
```

### Working with test data

```python
test_df = pd.read_csv("../datasets/housing-prices-dataset/test.csv",index_col='Id')
```

```python
test_df.head()
```

```python
test_df = test_df[list(set(test_df.columns)-set(object_columns))]
```

```python
test_df.head()
```

### Saving Both Datasets

```python
df.to_csv("../datasets/housing-prices-dataset/numerical_train.csv")
test_df.to_csv("../datasets/housing-prices-dataset/numerical_test.csv",index=False)
```
